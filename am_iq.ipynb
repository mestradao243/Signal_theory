# Señales AM y Demodulación: DSB-LC, DSB-SC, Envolvente, Coherente e IQ

Este desarrollo presenta la teoría matemática de las señales moduladas en amplitud (AM) y sus métodos de demodulación: detector de envolvente, demodulación coherente y demodulación IQ.

Se trabaja con frecuencia angular:

**ω = 2πf**

La señal de mensaje se denota como **x(t)**, la frecuencia angular de la portadora como **ωc**, y su fase como **θc**.

---

## Condición de la señal base x(t)

La señal de entrada se define como un **pulso cuadrado periódico**, aproximado mediante una serie de Fourier truncada.

### Forma de la señal:

x(t) = suma de k = 1, 3, 5 de: ak · cos(k ω0 t)

---

### Condiciones impuestas:

- Solo se consideran los **tres primeros armónicos impares**:
  - k = 1, 3, 5

- Se elimina el componente DC:
  - a0 = 0

- La señal tiene media cero:
  - promedio de x(t) = 0

---

### Serie de Fourier truncada (pulso cuadrado ideal):

x(t) ≈ (4/π) [  
(1/1)·cos(ω0 t)  
+ (1/3)·cos(3ω0 t)  
+ (1/5)·cos(5ω0 t)  
]

---

## Transformada de Fourier

Relación entre tiempo y frecuencia:

- X(ω): representa la señal en frecuencia
- x(t): representa la señal en el tiempo

(Transformada directa)  
X(ω) = integral de x(t) · e^(−jωt) dt

(Transformada inversa)  
x(t) = (1/2π) integral de X(ω) · e^(jωt) dω

---

## Interpretación en frecuencia

La señal x(t) contiene componentes en:

- ω0
- 3ω0
- 5ω0

y **no tiene componente en DC (ω = 0)**

---

## Interpretación física

- Señal tipo onda cuadrada simétrica
- Energía distribuida en armónicos impares
- Sin desplazamiento vertical (media cero)
- Modelo ideal para análisis de modulación

---

## Señales aleatorias (contexto general)

### Autocorrelación:
R_x(τ) = valor promedio de x(t) · x(t+τ)

### Densidad espectral de potencia (PSD):
S_x(ω) = distribución de potencia de la señal en frecuencia


## 1. Descripción en banda base, pasabanda y envolvente compleja

Una señal pasabanda real centrada en la frecuencia angular ωc puede representarse mediante una **envolvente compleja** u(t), asociada a la señal base definida previamente x(t), como:

s(t) = parte real de { u(t) · e^(j ωc t) }

---

## Relación con la señal definida

En este caso, la señal base es:

x(t) = suma de sus armónicos:
- ω0
- 3ω0
- 5ω0

sin componente DC (media cero)

Por lo tanto, la envolvente compleja u(t) puede interpretarse como una versión en banda base de x(t), es decir:

u(t) ≈ x(t) en representación compleja equivalente

---

## Descomposición en componentes en fase y cuadratura

La envolvente compleja se escribe como:

u(t) = I(t) + j Q(t)

donde:

- I(t): componente en fase (in-phase)
- Q(t): componente en cuadratura

---

## Señal pasabanda equivalente

Sustituyendo en la expresión general:

s(t) = I(t) cos(ωc t) − Q(t) sin(ωc t)

---

## En términos de la señal x(t)

Dado que x(t) es real y de media cero, construida como suma de cosenos:

x(t) ≈ suma de armónicos impares

entonces:

- x(t) puede verse como una realización en banda base de u(t)
- La modulación traslada estos armónicos alrededor de ±ωc

---

## Envolvente y fase instantánea

A partir de I(t) y Q(t), se definen:

### Envolvente:

A(t) = √( I²(t) + Q²(t) )

### Fase instantánea:

φ(t) = atan2( Q(t), I(t) )

---

## Forma equivalente de la señal pasabanda

La señal puede escribirse como:

s(t) = A(t) · cos(ωc t + φ(t))

---

## Interpretación física

- x(t) contiene la información en banda base (armónicos 1, 3 y 5 de ω0)
- u(t) representa esa información en forma compleja equivalente
- s(t) es la versión modulada en banda pasante alrededor de ωc
- La demodulación consiste en recuperar u(t) a partir de s(t)

---

## Interpretación en demodulación

Este modelo permite interpretar:

- Demodulación envolvente → estima A(t)
- Demodulación coherente → recupera I(t) y Q(t)
- Demodulación IQ → reconstruye completamente u(t)

---


## 2. Modulación AM de Doble Banda Lateral con Portadora Suprimida (DSB-SC)

En la modulación DSB-SC la portadora no se transmite como una línea espectral independiente. La señal transmitida se define como:

$$
s_{\mathrm{SC}}(t) = A_c \, m(t)\cos(\omega_c t)
$$

donde:
- $A_c$ es una constante de amplitud de la portadora
- $m(t)$ es la señal base, que en este caso es $x(t)$

---

## Relación con la señal definida

La señal base es:

$$
m(t) = x(t)
$$

y está definida como:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

con media cero:

$$
\mathbb{E}\{x(t)\} = 0
$$

---

## Representación de la portadora

La portadora se expresa como:

$$
\cos(\omega_c t) = \frac{1}{2}\left(e^{j\omega_c t} + e^{-j\omega_c t}\right)
$$

---

## Espectro de la señal DSB-SC

El espectro de la señal transmitida es:

$$
S_{\mathrm{SC}}(\omega)
= \frac{A_c}{2}\left[
M(\omega - \omega_c) + M(\omega + \omega_c)
\right]
$$

---

## Interpretación espectral

La señal base $x(t)$ se traslada a dos bandas laterales:

- Banda lateral superior (USB) alrededor de $+\omega_c$
- Banda lateral inferior (LSB) alrededor de $-\omega_c$

Si $x(t)$ está limitada en frecuencia a $|\omega| \leq W$, entonces:

$$
\omega_c - W \leq |\omega| \leq \omega_c + W
$$

---

## Ancho de banda

El ancho de banda total en la región positiva es:

$$
BW = 2W
$$

---

## Envolvente compleja

La envolvente compleja de la señal DSB-SC es:

$$
u_{\mathrm{SC}}(t) = A_c \, m(t)
$$

y como $m(t) = x(t)$:

$$
u_{\mathrm{SC}}(t) = A_c \, x(t)
$$

---

## Interpretación física

- La información está contenida en $x(t)$
- No existe portadora transmitida explícita
- La información se codifica como inversión de fase:
  - $x(t) > 0$ → fase $0$
  - $x(t) < 0$ → fase $\pi$

---

## Consecuencia en detección

Un detector de envolvente entrega:

$$
|u_{\mathrm{SC}}(t)| = A_c |x(t)|
$$

Esto implica:

- Se pierde el signo de $x(t)$
- Se introduce distorsión
- No es adecuado para DSB-SC

---




## 3. DSB-SC como proceso aleatorio

Se asume que la señal base $m(t)$ es un proceso aleatorio **estacionario en sentido amplio (WSS)**, de media cero, con:

- Función de autocorrelación: $R_m(\tau)$  
- Densidad espectral de potencia (PSD): $S_m(\omega)$  

---

## Señal modulada DSB-SC

La señal transmitida es:

$$
s_{\mathrm{SC}}(t) = A_c \, m(t)\cos(\omega_c t)
$$

En este caso, como $m(t) = x(t)$:

- $x(t)$ es el pulso cuadrado truncado con armónicos 1, 3 y 5
- tiene media cero
- es determinista en este desarrollo, pero se modela como proceso WSS para análisis estadístico

---

## Naturaleza del proceso

La multiplicación por la portadora $\cos(\omega_c t)$ introduce dependencia explícita del tiempo absoluto $t$, por lo que:

- $s_{\mathrm{SC}}(t)$ no es WSS
- es un proceso **ciclostacionario**

---

## Autocorrelación (convención de retardo simétrico)

La autocorrelación se define como:

$$
R_s(t,\tau) = \mathbb{E}\left\{
s\left(t+\frac{\tau}{2}\right)
s\left(t-\frac{\tau}{2}\right)
\right\}
$$

---

## Autocorrelación de DSB-SC

Sustituyendo la señal:

$$
\begin{aligned}
R_{s_{\mathrm{SC}}}(t,\tau)
&= A_c^2 \, R_m(\tau)
\cos\left(\omega_c\left(t+\frac{\tau}{2}\right)\right)
\cos\left(\omega_c\left(t-\frac{\tau}{2}\right)\right)
\end{aligned}
$$

Aplicando identidades trigonométricas:

$$
\begin{aligned}
R_{s_{\mathrm{SC}}}(t,\tau)
= \frac{A_c^2}{2} R_m(\tau)
\left[
\cos(\omega_c \tau) + \cos(2\omega_c t)
\right]
\end{aligned}
$$

---

## Promedio temporal (eliminación de dependencia en t)

Promediando sobre un período de la portadora:

$$
T_c = \frac{2\pi}{\omega_c}
$$

se obtiene:

$$
\overline{R}_{s_{\mathrm{SC}}}(\tau)
= \frac{1}{T_c}\int_0^{T_c} R_{s_{\mathrm{SC}}}(t,\tau)\,dt
$$

Resultado:

$$
\overline{R}_{s_{\mathrm{SC}}}(\tau)
= \frac{A_c^2}{2} R_m(\tau)\cos(\omega_c \tau)
$$

---

## Densidad espectral de potencia (PSD promedio)

La PSD promediada en ciclos es:

$$
\overline{S}_{s_{\mathrm{SC}}}(\omega)
=
\frac{A_c^2}{4}
\left[
S_m(\omega - \omega_c) + S_m(\omega + \omega_c)
\right]
$$

---

## Potencia promedio transmitida

La potencia promedio de la señal es:

$$
P_{\mathrm{SC}} = \overline{R}_{s_{\mathrm{SC}}}(0)
$$

Evaluando:

$$
P_{\mathrm{SC}} = \frac{A_c^2}{2} R_m(0)
$$

---

## Interpretación física

- El proceso base $x(t)$ (pulso cuadrado con armónicos 1, 3 y 5) es “estacionario” en el modelo estadístico
- La modulación traslada su contenido espectral a $\pm \omega_c$
- La señal resultante no es estacionaria debido a la portadora
- Solo al promediar sobre el período de la portadora se recupera una descripción estacionaria equivalente

---


## 4. Modulación AM de Doble Banda Lateral con Portadora (DSB-LC)

En la modulación AM convencional se añade una componente de portadora antes de modular la señal base:

$$
s_{\mathrm{LC}}(t) = A_c \left[1 + k_a m(t)\right]\cos(\omega_c t)
$$

donde:
- $A_c$ es la amplitud de la portadora
- $k_a$ es la sensibilidad de amplitud
- $m(t)$ corresponde a la señal base definida en este trabajo como $x(t)$

---

## Relación con la señal definida

En este caso:

$$
m(t) = x(t)
$$

donde:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

y cumple:

$$
\mathbb{E}\{x(t)\} = 0
$$

---

## Envolvente instantánea

La envolvente de la señal es:

$$
a(t) = A_c \left[1 + k_a m(t)\right]
$$

---

## Condición de no sobremodulación

Para evitar distorsión por inversión de envolvente:

$$
1 + k_a m(t) \geq 0 \quad \forall t
$$

Si:

$$
|m(t)| \leq m_{\max}
$$

entonces el índice de modulación es:

$$
\mu = k_a m_{\max}
$$

y la condición de operación sin sobremodulación es:

$$
0 \leq \mu \leq 1
$$

---

## Espectro de la señal DSB-LC

El espectro se expresa como:

$$
S_{\mathrm{LC}}(\omega)
=
\pi A_c \left[
\delta(\omega - \omega_c) + \delta(\omega + \omega_c)
\right]
+
\frac{A_c k_a}{2}
\left[
M(\omega - \omega_c) + M(\omega + \omega_c)
\right]
$$

---

## Interpretación espectral

La señal DSB-LC contiene:

- Una componente discreta de portadora en $\pm \omega_c$
- Dos copias desplazadas del espectro de $m(t)$ alrededor de $\pm \omega_c$

Como $m(t) = x(t)$, el espectro de la señal se compone de los armónicos en $\omega_0$, $3\omega_0$ y $5\omega_0$ trasladados alrededor de la portadora.

---

## Potencia promedio

Si $m(t)$ es de media cero con potencia promedio:

$$
P_m = \mathbb{E}\{m^2(t)\}
$$

entonces la potencia total transmitida es:

$$
P_{\mathrm{LC}} = \frac{A_c^2}{2}\left(1 + k_a^2 P_m\right)
$$

---

## Descomposición de potencia

Potencia de la portadora:

$$
P_c = \frac{A_c^2}{2}
$$

Potencia de las bandas laterales:

$$
P_{SB} = \frac{A_c^2 k_a^2 P_m}{2}
$$

---

## Interpretación física

- La información $x(t)$ modula la amplitud de una portadora existente
- Existe siempre una componente dominante en $\omega_c$
- La señal es adecuada para demodulación por envolvente
- Presenta menor eficiencia espectral que DSB-SC

---

## Conclusión

En DSB-LC, la señal base $x(t)$ (pulso cuadrado con armónicos 1, 3 y 5 sin componente DC) modula la amplitud de una portadora explícita, generando una señal donde la información está contenida en la envolvente de dicha portadora.

## 5. Señal analítica y envolvente

La señal analítica asociada a una señal pasabanda real $s(t)$ se define como:

$$
s_a(t) = s(t) + j\,\mathcal{H}\{s(t)\}
$$

donde $\mathcal{H}\{\cdot\}$ representa la transformada de Hilbert.

---

## Envolvente de la señal

La envolvente de la señal pasabanda se obtiene como el módulo de la señal analítica:

$$
A_s(t) = |s_a(t)|
$$

---

## Caso de una señal AM pasabanda

Si la señal tiene la forma:

$$
s(t) = a(t)\cos(\omega_c t)
$$

donde $a(t)$ es la envolvente lenta (banda base), en este trabajo se interpreta como la señal modulante $x(t)$ o una versión escalada de ella según el esquema AM (DSB-LC o DSB-SC).

---

## Condición de separación de Bedrosian

Si $a(t)$ es suficientemente de baja frecuencia respecto a $\omega_c$, se cumple aproximadamente:

$$
s_a(t) \approx a(t)e^{j\omega_c t}
$$

y por tanto:

$$
A_s(t) \approx |a(t)|
$$

---

## Interpretación con la señal definida x(t)

Dado que:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

y cumple que $\omega_0 \ll \omega_c$, entonces:

- $x(t)$ actúa como señal de baja frecuencia respecto a la portadora
- puede considerarse válida la aproximación de separación

---

## Implicación en demodulación por envolvente

La aproximación implica que:

- Para DSB-LC:
  $$
  A_s(t) \approx A_c \left[1 + k_a x(t)\right]
  $$
  por lo que la envolvente recupera directamente la información.

- Para DSB-SC:
  $$
  A_s(t) \approx A_c |x(t)|
  $$
  por lo que se pierde el signo de la señal original.

---

## Interpretación física

- La señal analítica separa la información en:
  - amplitud (envolvente)
  - fase (portadora compleja)
- La transformada de Hilbert permite construir una representación compleja equivalente de la señal real
- La envolvente es la magnitud de dicha representación compleja

---


## 6. Detección de envolvente para DSB-LC

Para la modulación DSB-LC, la señal transmitida es:

$$
s_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right]\cos(\omega_c t)
$$

donde $m(t)$ corresponde en este desarrollo a la señal $x(t)$ definida como pulso cuadrado con armónicos 1, 3 y 5 y media cero.

---

## Condición de no sobremodulación

Si se cumple la condición de no sobremodulación:

$$
1 + k_a m(t) \geq 0
$$

entonces la envolvente de la señal es:

$$
A_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right]
$$

---

## Recuperación mediante detector de envolvente

Un detector ideal de envolvente seguido de eliminación de componente DC produce:

$$
A_{\mathrm{LC}}(t) - A_c = A_c k_a m(t)
$$

por lo que la estimación de la señal base es:

$$
\widehat{m}(t) = \frac{A_{\mathrm{LC}}(t)-A_c}{A_c k_a} = m(t)
$$

---

## Interpretación con la señal x(t)

Como:

$$
m(t)=x(t)
$$

y:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

entonces la detección de envolvente permite recuperar exactamente esta combinación de armónicos siempre que no haya sobremodulación.

---

## Caso de sobremodulación

Si existe sobremodulación, ocurre que:

$$
1 + k_a m(t) < 0
$$

En este caso, la envolvente se convierte en:

$$
A_{\mathrm{LC}}(t)=A_c |1+k_a m(t)|
$$

---

## Consecuencia

Esto implica:

- Las partes negativas de la señal se reflejan hacia arriba
- Se introduce una distorsión no lineal
- La operación de valor absoluto destruye la información de signo
- La señal original $m(t)$ ya no puede recuperarse de forma única

---

## Interpretación física

- En condiciones normales, la envolvente sigue directamente a la señal $x(t)$ escalada y desplazada
- En sobremodulación, la envolvente deja de representar la señal original
- El detector de envolvente solo es válido en régimen lineal de modulación

---


## 7. Detección de envolvente para DSB-SC

Para la modulación DSB-SC, la señal transmitida es:

$$
s_{\mathrm{SC}}(t)=A_c\, m(t)\cos(\omega_c t)
$$

donde $m(t)$ corresponde en este desarrollo a la señal $x(t)$ definida como un pulso cuadrado con armónicos 1, 3 y 5, de media cero.

---

## Factor de baja frecuencia (envolvente equivalente)

El factor que modula la portadora es:

$$
a(t)=A_c\, m(t)
$$

Como $m(t)$ puede tomar valores positivos y negativos, la envolvente no puede seguir directamente la señal original.

---

## Envolvente obtenida

La envolvente del sistema (obtenida vía señal analítica o detector de magnitud) es aproximadamente:

$$
A_{\mathrm{SC}}(t)=A_c |m(t)|
$$

---

## Pérdida de información de signo

Dado que:

$$
A_{\mathrm{SC}}(t)=A_c |m(t)|
$$

se cumple que:

- Se pierde el signo de $m(t)$
- La señal original no puede recuperarse de forma única
- La operación se vuelve no invertible sin información adicional

---

## Interpretación con la señal definida x(t)

Como:

$$
m(t)=x(t)
$$

y:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

entonces la envolvente de DSB-SC corresponde al valor absoluto de esta suma de armónicos, lo que introduce distorsión estructural en la forma de onda.

---

## Interpretación estadística (caso general)

Si $m(t)$ es un proceso gaussiano de media cero con varianza $\sigma_m^2$, entonces $|m(t)|$ sigue una distribución half-normal en cada instante.

Sus momentos son:

### Media:

$$
\mathbb{E}\{|m(t)|\}=\sigma_m\sqrt{\frac{2}{\pi}}
$$

### Varianza:

$$
\operatorname{var}(|m(t)|)=\sigma_m^2\left(1-\frac{2}{\pi}\right)
$$

---

## Consecuencias en el dominio de la señal

La detección de envolvente en DSB-SC implica:

- Introducción de un componente DC positivo artificial
- Alteración de la estructura de segundo orden (autocorrelación y PSD)
- No preservación de la simetría de la señal original

---

## Interpretación física

- El detector de envolvente actúa como un operador de magnitud
- No distingue polaridad de la señal
- No es un demodulador coherente
- Introduce distorsión incluso en condiciones ideales

---


## Comportamiento de una señal bipolar en modulación LC y SC

Una señal bipolar es una señal $m(t)$ que puede tomar valores positivos y negativos. Este es el caso típico de señales de audio, procesos aleatorios de media cero y muchas señales de banda base en comunicaciones.

El problema fundamental en modulación es determinar qué ocurre cuando $m(t)$ cambia de signo.

---

## DSB-LC (portadora con nivel DC)

En DSB-LC la señal transmitida es:

$$
s_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right]\cos(\omega_c t)
$$

El término constante $1$ desplaza la señal bipolar hacia valores positivos antes de modularla.

---

## Interpretación del comportamiento

Si se cumple la condición de no sobremodulación:

$$
1+k_a m(t)\geq 0
$$

entonces:

- El factor que multiplica la portadora nunca se hace negativo
- La señal en pasabanda mantiene siempre la misma fase de portadora
- La información aparece como variaciones suaves de amplitud alrededor de $A_c$

En este caso:

- Si $m(t)>0$ → la envolvente aumenta por encima de $A_c$
- Si $m(t)<0$ → la envolvente disminuye pero permanece positiva

Por eso, un detector de envolvente puede recuperar la señal después de eliminar el nivel DC asociado a la portadora.

---

## DSB-SC (portadora suprimida)

En DSB-SC la señal transmitida es:

$$
s_{\mathrm{SC}}(t)=A_c\, m(t)\cos(\omega_c t)
$$

Aquí no existe desplazamiento positivo previo.

---

## Interpretación del cambio de signo

Cuando:

- $m(t)>0$ → la señal está en fase con la portadora
- $m(t)<0$ → ocurre una inversión de fase de $\pi$

Esto se puede escribir como:

$$
A_c m(t)\cos(\omega_c t)
=
A_c |m(t)|\cos(\omega_c t + \pi)
$$

---

## Consecuencia física

- El cambio de signo en $m(t)$ no aparece en la amplitud
- Aparece como un cambio de fase de $\pi$
- La envolvente física es:

$$
A_{\mathrm{SC}}(t)=A_c |m(t)|
$$

Por lo tanto, la envolvente no contiene información de signo.

---

## Interpretación con la señal definida x(t)

En este trabajo:

$$
m(t)=x(t)
$$

y:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

Como esta señal es bipolar (media cero), cruza frecuentemente por cero, lo que en DSB-SC se traduce directamente en inversiones de fase en la portadora.

---

## Diferencia fundamental entre LC y SC

- En DSB-LC:
  - La bipolaridad se convierte en variación de amplitud alrededor de un nivel positivo
  - La fase de la portadora se mantiene constante
  - La información está en la envolvente

- En DSB-SC:
  - La bipolaridad se convierte en cambios de fase de $\pi$
  - La amplitud es siempre positiva en valor absoluto
  - La información de signo no está en la envolvente

---

## Conclusión

DSB-LC y DSB-SC representan de forma distinta una misma señal bipolar $x(t)$:

- DSB-LC la transforma en variaciones de amplitud sobre una portadora existente
- DSB-SC la codifica como cambios de fase de la portadora

En consecuencia, LC permite demodulación por envolvente, mientras que SC requiere demodulación coherente o IQ para recuperar completamente la información original.

## 8. Demodulación coherente

La demodulación coherente consiste en multiplicar la señal pasabanda recibida por una portadora local sincronizada y luego aplicar un filtro pasa-bajas para eliminar los términos de alta frecuencia.

Se define la señal recibida como:

$$
r(t)=s(t)\cos(\omega_c t+\phi)
$$

donde $\phi$ representa el error de fase entre el oscilador del receptor y la portadora transmitida.

---

## Caso DSB-SC con tu señal \(x(t)\)

En este trabajo:

$$
m(t)=x(t)
$$

y:

$$
s_{\mathrm{SC}}(t)=A_c x(t)\cos(\omega_c t)
$$

---

## Multiplicación coherente

Al multiplicar:

$$
\begin{aligned}
r_{\mathrm{SC}}(t)
&= A_c x(t)\cos(\omega_c t)\cos(\omega_c t+\phi)
\end{aligned}
$$

Usando identidades trigonométricas:

$$
\begin{aligned}
r_{\mathrm{SC}}(t)
&= \frac{A_c}{2}x(t)\cos(\phi)
+ \frac{A_c}{2}x(t)\cos(2\omega_c t+\phi)
\end{aligned}
$$

---

## Filtrado pasa-bajas

Aplicando un filtro ideal pasa-bajas que elimina el término en $2\omega_c$:

$$
y_{\mathrm{SC}}(t)=\mathrm{LPF}\{2r_{\mathrm{SC}}(t)\}
$$

se obtiene:

$$
y_{\mathrm{SC}}(t)=A_c x(t)\cos(\phi)
$$

---

## Interpretación del factor de fase

El resultado depende críticamente de $\phi$:

- Si $\phi=0$:
  $$
  y_{\mathrm{SC}}(t)=A_c x(t)
  $$
  recuperación perfecta (salvo escala)

- Si $\phi=\pi/2$:
  $$
  y_{\mathrm{SC}}(t)=0
  $$
  pérdida total de la señal

- Si $\phi=\pi$:
  $$
  y_{\mathrm{SC}}(t)=-A_c x(t)
  $$
  inversión de signo

---

## Caso DSB-LC con tu señal \(x(t)\)

La señal es:

$$
s_{\mathrm{LC}}(t)=A_c\left[1+k_a x(t)\right]\cos(\omega_c t)
$$

---

## Demodulación coherente

Multiplicando por el oscilador local:

$$
s_{\mathrm{LC}}(t)\cos(\omega_c t+\phi)
$$

y aplicando filtrado pasa-bajas:

$$
y_{\mathrm{LC}}(t)=\mathrm{LPF}\{2s_{\mathrm{LC}}(t)\cos(\omega_c t+\phi)\}
$$

---

## Resultado

Se obtiene:

$$
y_{\mathrm{LC}}(t)=A_c\left[1+k_a x(t)\right]\cos(\phi)
$$

---

## Eliminación de componente DC

Separando el término constante:

$$
y_{\mathrm{LC}}(t)=A_c\cos(\phi)+A_c k_a x(t)\cos(\phi)
$$

Restando el DC y reescalando:

$$
\hat{x}(t)=x(t)
$$

siempre que:

$$
\cos(\phi)\neq 0
$$

---

## Interpretación física

- La demodulación coherente recupera la información de $x(t)$ usando referencia de fase
- En DSB-SC es crítica la sincronización de fase
- En DSB-LC el DC facilita la recuperación de la envolvente antes del escalamiento
- El error de fase introduce atenuación o inversión de la señal

---



## 9. Demodulación coherente en el dominio de la frecuencia

Para DSB-SC, la multiplicación por el oscilador local sincronizado genera traslación espectral hacia banda base y hacia alta frecuencia.

En este trabajo, la señal base es:

$$
m(t)=x(t)
$$

donde:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

---

## Multiplicación coherente

Al multiplicar la señal DSB-SC por el oscilador local:

$$
2s_{\mathrm{SC}}(t)\cos(\omega_c t+\phi)
$$

se obtiene:

$$
A_c x(t)\cos(\phi) + A_c x(t)\cos(2\omega_c t+\phi)
$$

---

## Interpretación espectral

Este resultado tiene dos componentes:

- Componente en banda base: centrada en $\omega \approx 0$
- Imagen de alta frecuencia: centrada en $\pm 2\omega_c$

---

## Condición del filtro pasa-bajas

El filtro debe cumplir:

$$
W < \omega_{\mathrm{LPF}} < 2\omega_c - W
$$

donde $W$ es el ancho de banda de $x(t)$.

Dado que $x(t)$ contiene armónicos en:

- $\omega_0$
- $3\omega_0$
- $5\omega_0$

su ancho de banda efectivo es aproximadamente $W \approx 5\omega_0$.

---

## Señal de salida en banda base

Después del filtrado ideal:

$$
y(t)=A_c x(t)\cos(\phi)
$$

---

## Densidad espectral de potencia (PSD)

Si $x(t)$ es un proceso WSS de media cero, la PSD de salida es:

$$
S_y(\omega)=A_c^2\cos^2(\phi)\,S_x(\omega)
$$

---

## Autocorrelación

La autocorrelación resultante es:

$$
R_y(\tau)=A_c^2\cos^2(\phi)\,R_x(\tau)
$$

---

## Interpretación con la señal definida

Como $x(t)$ es un pulso cuadrado truncado con armónicos:

- la estructura espectral (líneas en $\omega_0$, $3\omega_0$, $5\omega_0$)
- se preserva completamente en el receptor

solo escalada por $\cos(\phi)$.

---

## Efecto del error de fase

- Si $\phi=0$:
  $$
  y(t)=A_c x(t)
  $$

- Si $\phi=\pi/2$:
  $$
  y(t)=0
  $$

- Si $\phi=\pi$:
  $$
  y(t)=-A_c x(t)
  $$

---


## 10. Demodulación IQ

La demodulación IQ permite estimar simultáneamente las proyecciones en fase y en cuadratura de una señal pasabanda.

En este desarrollo, la señal base es:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

y se utiliza como componente de banda base dentro de la envolvente compleja del sistema.

---

## Proyecciones en fase y cuadratura

Se definen las componentes IQ como:

$$
I(t)=\operatorname{LPF}\{2s(t)\cos(\omega_c t)\}
$$

$$
Q(t)=\operatorname{LPF}\{-2s(t)\sin(\omega_c t)\}
$$

donde $s(t)$ es la señal pasabanda modulada a partir de $x(t)$.

---

## Representación en envolvente compleja

Una señal pasabanda puede expresarse como:

$$
s(t)=\operatorname{Re}\{u(t)e^{j\omega_c t}\}
$$

donde la envolvente compleja es:

$$
u(t)=I_u(t)+jQ_u(t)
$$

---

## Recuperación ideal mediante demodulación IQ

Bajo condiciones ideales de filtrado pasa-bajas (eliminando componentes alrededor de $2\omega_c$):

$$
I(t)=I_u(t),
\qquad
Q(t)=Q_u(t)
$$

---

## Interpretación con la señal x(t)

Dado que la señal base es real y de media cero:

- $x(t)$ no introduce componente imaginaria por sí misma
- actúa como la parte en fase de la envolvente compleja en un sistema ideal alineado

Por lo tanto, en un caso perfectamente sincronizado:

$$
u(t)\approx x(t)
$$

y la información contenida en los armónicos en:

- $\omega_0$
- $3\omega_0$
- $5\omega_0$

se preserva completamente en la reconstrucción IQ.

---

## Forma equivalente con mezcla compleja

La demodulación IQ puede expresarse como:

$$
\operatorname{LPF}\{2s(t)e^{-j\omega_c t}\}=u(t)
$$

---

## Interpretación espectral

La multiplicación por $e^{-j\omega_c t}$ traslada el espectro de:

- alrededor de $+\omega_c$
- hacia banda base ($\omega \approx 0$)

mientras el filtro elimina las imágenes alrededor de $2\omega_c$.

---

## Ventajas de IQ

- Recupera simultáneamente amplitud y fase
- No pierde el signo de la señal base $x(t)$
- Permite reconstrucción completa de la envolvente compleja
- Es robusta frente a cambios de fase del portador (con estimación adecuada)

---


## 11. Vista IQ de DSB-SC y DSB-LC

---

## DSB-SC en fase alineada

Para la modulación DSB-SC, la señal es:

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t)
=\operatorname{Re}\{A_c m(t)e^{j\omega_c t}\}
$$

---

## Envolvente compleja e IQ

De esta forma:

$$
u_{\mathrm{SC}}(t)=A_c m(t)
$$

y las componentes IQ ideales son:

$$
I(t)=A_c m(t)
$$

$$
Q(t)=0
$$

---

## Caso con error de fase constante

Si existe un desfase $\theta$ en la portadora:

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t+\theta)
=\operatorname{Re}\{A_c m(t)e^{j\theta}e^{j\omega_c t}\}
$$

---

## Salida IQ con error de fase

Las componentes quedan:

$$
I(t)=A_c m(t)\cos\theta
$$

$$
Q(t)=A_c m(t)\sin\theta
$$

---

## Recuperación de la señal base

La señal original se recupera proyectando sobre la fase correcta:

$$
\widehat{m}(t)=\frac{I(t)\cos\theta + Q(t)\sin\theta}{A_c}
$$

---

## Magnitud IQ

La magnitud del vector IQ es:

$$
\sqrt{I^2(t)+Q^2(t)}=A_c|m(t)|
$$

Esto implica:

- Se pierde la información de signo
- El sistema se comporta como un detector de envolvente en magnitud
- El signo de $m(t)$ está contenido en la fase del número complejo

---

## Interpretación con la señal definida x(t)

En este trabajo:

$$
m(t)=x(t)
$$

donde:

$$
x(t) \approx \frac{4}{\pi}\left[
\cos(\omega_0 t)
+ \frac{1}{3}\cos(3\omega_0 t)
+ \frac{1}{5}\cos(5\omega_0 t)
\right]
$$

Por lo tanto:

- los cruces por cero de $x(t)$ se reflejan como saltos de fase de $\pi$
- la estructura armónica se preserva en la representación IQ compleja

---

## DSB-LC en vista IQ

Bajo no sobremodulación:

$$
u_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right]
$$

---

## Componentes IQ

En fase alineada:

$$
I(t)=A_c\left[1+k_a m(t)\right]
$$

$$
Q(t)=0
$$

---

## Interpretación física

- En DSB-SC:
  - la información está en la fase del vector IQ
  - el signo de $x(t)$ se representa como cambios de fase

- En DSB-LC:
  - existe un offset DC dominante en $I(t)$
  - la información está en variaciones alrededor de ese nivel

---


## 12. Comparación entre receptores de envolvente, coherente e IQ

### DSB-LC (portadora con nivel DC)

- **Detector de envolvente**
  
  Salida ideal:
  $$
  A_c[1+k_a m(t)]
  $$
  
  Condición de validez:
  $$
  1+k_a m(t)\geq 0
  $$
  
  Falla principal:
  - sobremodulación (la envolvente se invierte y deja de ser invertible)

---

- **Demodulación coherente**
  
  Salida ideal:
  $$
  A_c[1+k_a m(t)]\cos\phi
  $$
  
  Condición:
  - sincronización en frecuencia y fase
  - $\cos\phi \neq 0$
  
  Falla principal:
  - error de fase (atenuación o anulación en cuadratura)

---

- **Demodulación IQ**
  
  Salida ideal:
  $$
  A_c[1+k_a m(t)]e^{j\theta}
  $$
  
  Condición:
  - separación ideal de banda base respecto a imágenes en $2\omega_c$
  
  Falla principal:
  - error en estimación de portadora (fase o frecuencia)

---

### DSB-SC (portadora suprimida)

- **Detector de envolvente**
  
  Salida:
  $$
  A_c |m(t)|
  $$
  
  Condición:
  - solo útil si no se necesita información de signo
  
  Falla principal:
  - pérdida completa de la polaridad de $m(t)$

---

- **Demodulación coherente**
  
  Salida ideal:
  $$
  A_c m(t)\cos\phi
  $$
  
  Condición:
  - sincronización de portadora
  
  Falla principal:
  - error de fase o cancelación en cuadratura

---

- **Demodulación IQ**
  
  Salida ideal:
  $$
  A_c m(t)e^{j\theta}
  $$
  
  Condición:
  - referencia compleja de portadora disponible o estimable
  
  Falla principal:
  - si se usa solo magnitud, se pierde el signo de $m(t)$

---

## Interpretación unificada

La diferencia entre los receptores puede entenderse así:

- **Detector de envolvente:** operación de magnitud (pierde signo)
- **Demodulación coherente:** proyección sobre una portadora conocida (recupera signo si hay sincronía)
- **Demodulación IQ:** reconstrucción completa de la envolvente compleja

---

## Diferencia fundamental entre DSB-LC y DSB-SC

### DSB-LC

La señal es:

$$
s_{\mathrm{LC}}(t)=A_c[1+k_a m(t)]\cos(\omega_c t)
$$

- Existe un término constante “1” que actúa como referencia de portadora
- La información se codifica como variaciones de amplitud alrededor de un nivel positivo
- La envolvente es directamente interpretable físicamente

---

### DSB-SC

La señal es:

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t)
$$

- No existe portadora transmitida explícita
- El cambio de signo de $m(t)$ se traduce en cambios de fase de $\pi$
- La información está en la fase, no en la magnitud

---

## Conclusión

- DSB-LC permite detección por envolvente porque conserva una referencia positiva en amplitud.
- DSB-SC elimina la portadora y codifica la información en fase, por lo que requiere demodulación coherente o IQ.
- IQ es la representación más completa, ya que preserva la estructura compleja de la señal base $x(t)$ en ambos esquemas.


## Estructura del experimento por etapas

El experimento asignado se organiza en cuatro etapas:

I. **Señal de entrada**: definición del pulso rectangular y su convención de muestreo.

II. **Modulación DSB**: generación de señales pasabanda DSB-LC y DSB-SC para los índices de modulación solicitados.

III. **Canal AWGN**: adición de ruido blanco gaussiano en pasabanda para las relaciones señal-ruido (SNR) requeridas.

IV. **Demodulación**: recuperación de la señal mediante detectores de envolvente, demodulación coherente y demodulación IQ; posteriormente se comparan forma de onda, función de autocorrelación (ACF), densidad espectral de potencia (PSD) y comportamiento del ancho de banda.


## I. Señal de entrada

El estudio empírico utiliza un pulso rectangular finito como señal de entrada. El número de muestras es:

$$
N = 4096
$$

y el incremento de muestreo solicitado es:

$$
\Delta t = \frac{1}{N}
$$

Por lo tanto, la frecuencia de muestreo equivalente para los cálculos en el dominio de la frecuencia es:

$$
f_s = \frac{1}{\Delta t} = N = 4096
$$

---

## Definición de la señal discreta

La señal de mensaje en tiempo discreto es un pulso rectangular unipolar de amplitud 1 y duración de 2048 muestras:

$$
m[n] =
\begin{cases}
1, & 0 \leq n < 2048 \\
0, & 2048 \leq n < 4096
\end{cases}
$$

---

## Propiedades estadísticas básicas

El ciclo de trabajo es:

$$
D = \frac{1}{2}
$$

por lo que su media discreta es:

$$
\mathbb{E}\{m[n]\} = \frac{1}{2}
$$

---

## Descomposición en componente DC y componente alterna

La señal puede separarse como:

- Componente DC:
  $$
  \frac{1}{2}
  $$

- Componente de información:
  $$
  m[n] - \frac{1}{2}
  $$

La componente centrada contiene la transición del pulso, mientras que la componente DC genera una línea espectral en frecuencia cero.

---

## Interpretación espectral

Debido a las transiciones abruptas del pulso rectangular:

- el espectro presenta forma tipo sinc
- aparecen lóbulos secundarios (sidelobes)
- existe contenido significativo en altas frecuencias

---

## Interpretación periódica del experimento

Para el análisis empírico, el registro completo (alto + bajo) se interpreta como un período de una señal periódica rectangular.

Por lo tanto, la frecuencia fundamental del mensaje es:

$$
f_m = \frac{1}{N \Delta t}
$$

---

## Selección de la portadora

La frecuencia de la portadora se define como:

$$
f_c = 30 f_m
$$

---

## Interpretación con el modelo del sistema

En relación con el resto del desarrollo:

- esta señal reemplaza a la señal ideal continua $x(t)$ usada en las secciones teóricas
- introduce efectos reales de discretización y espectro sinc
- mantiene la condición de separación de banda respecto a la portadora

---

## Conclusión

El sistema utiliza un pulso rectangular discreto como aproximación de la señal base del sistema de comunicaciones. Su estructura de media no cero, transiciones abruptas y contenido espectral amplio permite evaluar el comportamiento de la modulación DSB bajo condiciones más realistas.


## II. Modulación DSB

---

## DSB-LC (portadora con nivel DC)

La modulación DSB-LC se implementa variando el índice de modulación como:

$$
s_{\mathrm{LC},\mu}[n]=A_c\left[1+\mu m[n]\right]\cos(2\pi f_c n\Delta t),
\qquad
\mu \in \{0.72,\,1.00,\,1.23\}
$$

---

## Relación con la señal definida

En este experimento:

$$
m[n] =
\begin{cases}
1, & 0 \leq n < 2048 \\
0, & 2048 \leq n < 4096
\end{cases}
$$

Por lo tanto:

$$
m[n] \in \{0,1\}
$$

---

## Interpretación del índice de modulación

El factor de envolvente es:

$$
1 + \mu m[n]
$$

lo cual implica:

- cuando $m[n]=0$:
  $$
  1+\mu m[n] = 1
  $$

- cuando $m[n]=1$:
  $$
  1+\mu m[n] = 1+\mu
  $$

---

## Observación importante del experimento

Dado que $m[n]$ es unipolar:

- no existe cambio de signo en la señal modulante
- el sistema no entra en sobremodulación clásica
- los valores $\mu \in \{0.72, 1.00, 1.23\}$ representan únicamente diferentes niveles de profundidad de modulación

---

## Interpretación física en DSB-LC

- el nivel base de la portadora siempre está presente
- la información se codifica como variación entre dos niveles de amplitud
- el espectro de banda lateral crece con $\mu$

---

## DSB-SC (portadora suprimida)

La señal transmitida es:

$$
s_{\mathrm{SC}}[n]=A_c\, m[n]\cos(2\pi f_c n\Delta t)
$$

---

## Interpretación con señal unipolar

Como:

$$
m[n] \in \{0,1\}
$$

entonces:

- cuando $m[n]=1$ → la portadora está activa
- cuando $m[n]=0$ → la portadora se anula

---

## Envolvente en DSB-SC

En el caso ideal sin ruido:

$$
A_{\mathrm{SC}}[n] = A_c m[n]
$$

lo que implica:

- la envolvente reproduce exactamente la forma del pulso rectangular
- no hay pérdida de signo porque la señal nunca es negativa

---

## Diferencia clave con caso bipolar

En una señal bipolar:

- la envolvente sería $A_c |m[n]|$
- se perdería información de signo

En este caso unipolar:

- no hay ambigüedad de signo
- la detección por envolvente es equivalente a la señal original escalada

---

## Interpretación espectral del efecto de μ

En DSB-LC:

- al aumentar $\mu$:
  - aumenta la potencia en bandas laterales
  - aumenta la variación de amplitud alrededor de la portadora

En DSB-SC:

- la energía transmitida depende directamente de la actividad de $m[n]$
- no existe componente portadora independiente

---

## Conclusión

En este experimento, la señal rectangular unipolar transforma la modulación DSB en un caso particular donde:

- DSB-LC controla la amplitud entre dos niveles definidos por $1$ y $1+\mu$
- DSB-SC actúa como una conmutación de portadora (on/off) sincronizada con $m[n]$

Esto permite analizar de forma directa el impacto del índice de modulación sin los efectos de signo presentes en señales bipolares.

## III. Canal AWGN

El canal añade ruido blanco gaussiano en banda pasante.

Para una relación señal a ruido objetivo en dB:

$$
\sigma_n^2=\frac{P_s}{10^{\mathrm{SNR}_{\mathrm{dB}}/10}}
$$

donde:
- $P_s$ es la potencia empírica de la señal en banda pasante
- $\sigma_n^2$ es la varianza del ruido

---

## Señal recibida

La señal en el receptor se modela como:

$$
r[n]=s[n]+n[n]
$$

donde:
- $s[n]$ es la señal modulada (DSB-LC o DSB-SC)
- $n[n]$ es ruido blanco gaussiano aditivo

---

## Valores de SNR utilizados

El experimento evalúa tres condiciones de canal:

$$
\mathrm{SNR}_{\mathrm{dB}} \in \{30, 10, 3\}
$$

---

## Interpretación del canal

- A 30 dB:
  - el ruido es bajo
  - la estructura de la señal se conserva claramente

- A 10 dB:
  - aparecen distorsiones visibles en forma de onda
  - el espectro se ensancha

- A 3 dB:
  - el ruido domina parcialmente la señal
  - la recuperación depende fuertemente del tipo de demodulación

---

## Evaluación del efecto del canal

El impacto del ruido se analiza en dos niveles:

### 1. Nivel pasabanda

Se estudia directamente sobre $r[n]$:

- función de autocorrelación (ACF)
- densidad espectral de potencia (PSD)
- estimación de ancho de banda

---

### 2. Nivel de demodulación

Se evalúa cómo el ruido afecta:

- detector de envolvente
- demodulación coherente
- demodulación IQ

---

## Interpretación física

El canal AWGN:

- no altera la estructura determinista de la modulación
- introduce variaciones aleatorias en amplitud y fase
- degrada progresivamente la recuperación de $m[n] = x[n]$

---

## Conclusión

El canal AWGN permite analizar la robustez de los esquemas DSB-LC y DSB-SC frente a ruido aditivo, evaluando cómo la degradación en banda pasante se traduce en errores en cada tipo de demodulación.

## IV. Demodulación

Los demoduladores se evalúan de la siguiente manera.

---

## 1. Detector de envolvente

Se utiliza un detector de pico basado en diodo con descarga RC.

- **Caso de buena inercia:**
  - La constante de tiempo RC es adecuada
  - Se suprime el rizado de la portadora
  - La envolvente sigue correctamente los bordes del pulso rectangular $m[n]$

- **Caso de mala inercia:**
  - La constante de descarga es demasiado lenta
  - La envolvente decae de forma retardada después del flanco de bajada
  - Se produce distorsión temporal en la recuperación

---

## 2. Detector coherente

El receptor realiza multiplicación con una portadora local sincronizada:

$$
\widehat{x}_{\phi}[n]=\operatorname{LPF}\{2r[n]\cos(2\pi f_c n\Delta t+\phi)\}
$$

---

## Interpretación en este experimento

Dado que:

- $m[n]$ es la señal base (pulso rectangular unipolar)
- $s[n]$ es DSB-LC o DSB-SC
- $r[n]=s[n]+n[n]$ (canal AWGN)

---

## Salida ideal sin ruido

### Para DSB-SC:

$$
\widehat{x}_{\phi}[n]=A_c m[n]\cos(\phi)
$$

### Para DSB-LC:

$$
\widehat{x}_{\phi}[n]=A_c\left[1+\mu m[n]\right]\cos(\phi)
$$

---

## Barrido de fase

El experimento utiliza:

$$
\phi \in \left\{0,\frac{\pi}{16},\frac{2\pi}{16},\ldots,\frac{8\pi}{16}\right\}
$$

---

## Interpretación del error de fase

- $\phi = 0$:
  - recuperación óptima (máxima correlación con $m[n]$)

- $\phi \neq 0$:
  - atenuación proporcional a $\cos(\phi)$
  - pérdida progresiva de energía recuperada

- $\phi = \pi/2$:
  - anulación total de la señal recuperada

---

## 3. Demodulador IQ

La demodulación compleja estima la envolvente compleja:

$$
\widehat{u}[n]=\operatorname{LPF}\{2r[n]e^{-j(2\pi f_c n\Delta t+\phi)}\}
$$

---

## Interpretación

- El resultado es una representación compleja de la señal base
- El error de fase produce una rotación en el plano complejo
- Si la fase es conocida o estimada, la rotación puede corregirse

---

## IQ imbalance

Se modela mediante:

- ganancias desiguales en I y Q
- pequeño desajuste de fase en cuadratura

Esto provoca:

- distorsión geométrica del plano complejo
- deformación del diagrama de constelación equivalente
- errores residuales incluso tras corrección de fase

---

## Métricas de evaluación

El desempeño de los demoduladores se evalúa mediante:

- correlación con la señal original $m[n]$
- error cuadrático medio normalizado (NMSE)
- SNR recuperada en banda base

---

## Ajuste afín

En algunas comparaciones se aplica un ajuste lineal óptimo:

$$
\widehat{m}_{\text{fit}}[n] = a\,\widehat{m}[n] + b
$$

Esto permite separar:

- errores de forma (distorsión)
- errores de escala (ganancia)
- errores de offset (DC residual)

---

## Conclusión

La etapa de demodulación permite comparar directamente la capacidad de cada receptor para reconstruir la señal base $m[n]$ (pulso rectangular unipolar) bajo condiciones de ruido AWGN, error de fase y desajustes del receptor. Cada técnica presenta un compromiso distinto entre robustez, complejidad y sensibilidad a sincronización.

## Representaciones cuadratura y canónica para LC y SC

Para toda señal pasabanda real centrada en $f_c$, existen dos descripciones equivalentes.

---

## 1. Representación en cuadratura

La señal puede escribirse como:

$$
s[n]=I[n]\cos(2\pi f_c n\Delta t)-Q[n]\sin(2\pi f_c n\Delta t)
$$

donde la envolvente compleja asociada es:

$$
u[n]=I[n]+jQ[n]
$$

---

## 2. Representación canónica amplitud–fase

La misma señal puede expresarse como:

$$
s[n]=A[n]\cos(2\pi f_c n\Delta t+\varphi[n])
$$

donde:

$$
A[n]=\sqrt{I^2[n]+Q^2[n]}
$$

$$
\varphi[n]=\operatorname{atan2}(Q[n],I[n])
$$

---

# DSB-LC

Para la señal definida en este experimento:

$$
s_{\mathrm{LC},\mu}[n]=A_c\left[1+\mu m[n]\right]\cos(2\pi f_c n\Delta t)
$$

y recordando que:

$$
m[n]=
\begin{cases}
1, & 0 \leq n < 2048 \\
0, & 2048 \leq n < 4096
\end{cases}
\quad \Rightarrow \quad m[n]\in\{0,1\}
$$

---

## Componentes IQ ideales

$$
I_{\mathrm{LC}}[n]=A_c\left[1+\mu m[n]\right]
$$

$$
Q_{\mathrm{LC}}[n]=0
$$

---

## Representación canónica

Como $I_{\mathrm{LC}}[n] > 0$ siempre:

$$
A_{\mathrm{LC}}[n]=A_c\left[1+\mu m[n]\right]
$$

$$
\varphi_{\mathrm{LC}}[n]=0
$$

---

## Interpretación física

- La información se codifica como variación de amplitud positiva
- Existe un nivel base de portadora siempre presente
- No hay ambigüedad de fase en este caso unipolar

---

# DSB-SC

La señal es:

$$
s_{\mathrm{SC}}[n]=A_c m[n]\cos(2\pi f_c n\Delta t)
$$

---

## Componentes IQ ideales

$$
I_{\mathrm{SC}}[n]=A_c m[n]
$$

$$
Q_{\mathrm{SC}}[n]=0
$$

---

## Representación canónica

Dado que $m[n]\in\{0,1\}$:

$$
A_{\mathrm{SC}}[n]=A_c m[n]
$$

$$
\varphi_{\mathrm{SC}}[n]=0 \quad \text{cuando } m[n]=1
$$

y

- cuando $m[n]=0$, la amplitud es cero y la fase no está definida físicamente.

---

## Nota sobre el caso general

Si el mensaje fuera bipolar:

- $m[n] < 0$ produciría:
  - $A_{\mathrm{SC}}[n]=A_c |m[n]|$
  - saltos de fase de $\pi$

---

## Estimación IQ en el experimento

El demodulador IQ produce:

$$
\widehat{u}[n]=\widehat{I}[n]+j\widehat{Q}[n]
$$

A partir de esto se calculan:

$$
(\widehat{I}[n],\widehat{Q}[n])
$$

$$
\widehat{A}[n]=|\widehat{u}[n]|
$$

$$
\widehat{\varphi}[n]=\operatorname{unwrap}(\arg \widehat{u}[n])
$$

---

## Interpretación de la comparación numérica

El análisis compara:

- representación LC vs SC en el plano IQ
- efecto del ruido AWGN
- efecto de desbalance IQ (ganancia y fase)
- estabilidad de amplitud y fase recuperadas

---

## Conclusión

Las representaciones en cuadratura y canónica muestran que tanto DSB-LC como DSB-SC pueden interpretarse como casos particulares de una misma estructura compleja. En este experimento, la señal unipolar $m[n]$ hace que:

- LC codifique información en amplitud sobre un nivel base positivo
- SC codifique información como activación/desactivación de la portadora

La representación IQ permite unificar ambos casos en el dominio complejo de la envolvente.


## Etapa por etapas: análisis ACF, PSD y ancho de banda

Esta sección hace explícito el análisis de segundo orden para cada etapa de la cadena de comunicaciones del experimento. La señal de entrada es el pulso rectangular finito:

$$
m[n]=u[n]-u[n-2048],
\qquad
0\leq n<4096
$$

con:
- amplitud 1  
- duración $L=2048$ muestras  
- incremento de muestreo $\Delta t = 1/4096$

---

# 1. Etapa de entrada

## Autocorrelación empírica

Para la secuencia determinista finita, la autocorrelación estimada es:

$$
\widehat{R}_m[\ell]=\frac{1}{N-\ell}\sum_{n=0}^{N-1-\ell}(m[n]-\bar m)(m[n+\ell]-\bar m),
\qquad
\ell \geq 0
$$

---

## Interpretación espectral

Si se analiza el pulso rectangular sin centrar, su transformada tiene forma tipo Dirichlet:

$$
M(e^{j\Omega})=
e^{-j\Omega(L-1)/2}
\frac{\sin(\Omega L/2)}{\sin(\Omega/2)},
\qquad
L=2048
$$

---

## PSD del pulso

La densidad espectral de potencia presenta:

- componente DC dominante (por ser unipolar)
- lóbulo principal controlado por la duración del pulso
- lóbulos secundarios por discontinuidades en los bordes

En implementación empírica, la PSD suele calcularse tras remover la media, de modo que el ancho de banda reportado corresponde al contenido AC de la señal.

---

# 2. Etapa DSB-SC

La señal es:

$$
s_{\mathrm{SC}}[n]=A_c m[n]\cos(2\pi f_c n\Delta t)
$$

---

## PSD promedio

$$
\overline{S}_{s_{\mathrm{SC}}}(f)=\frac{A_c^2}{4}
\left[
S_m(f-f_c)+S_m(f+f_c)
\right]
$$

---

## Interpretación

- el espectro del mensaje se traslada a $\pm f_c$
- la forma del espectro se conserva
- el ancho de banda está determinado por la transición del pulso rectangular

---

# 3. Etapa DSB-LC

La señal es:

$$
s_{\mathrm{LC}}[n]=A_c[1+\mu m[n]]\cos(2\pi f_c n\Delta t)
$$

---

## PSD promedio

$$
\overline{S}_{s_{\mathrm{LC}}}(f)=
\frac{A_c^2}{4}\left[\delta(f-f_c)+\delta(f+f_c)\right]
+
\frac{A_c^2\mu^2}{4}\left[
S_m(f-f_c)+S_m(f+f_c)
\right]
$$

---

## Interpretación

- existe una línea espectral en la portadora
- aparecen bandas laterales idénticas a las de DSB-SC
- el parámetro $\mu$ controla la potencia en las bandas laterales

Dado que $m[n]\in\{0,1\}$:

- el término $1+\mu m[n]$ siempre es positivo
- no hay inversión de envolvente

---

# 4. Canal AWGN

La señal recibida es:

$$
r[n]=s[n]+w[n]
$$

donde $w[n]$ es ruido blanco gaussiano independiente.

---

## Relación de segundo orden

$$
R_r[\ell]=R_s[\ell]+R_w[\ell]
$$

$$
S_r(f)=S_s(f)+S_w(f)
$$

---

## Interpretación

- el ruido añade energía en todo el espectro
- incrementa el ancho de banda aparente en estimaciones empíricas
- degrada la estructura del ACF del mensaje

---

# 5. Detector de envolvente

## Caso LC (inercia adecuada)

$$
\widehat{m}_{\mathrm{env}}[n]\approx m[n]
$$

---

## Caso SC (señal unipolar)

En condiciones ideales:

- la envolvente reproduce correctamente el soporte del pulso
- no hay ambigüedad de signo porque $m[n]\in\{0,1\}$

---

## Caso mala inercia

- la descarga RC lenta introduce alargamiento del pulso
- se distorsiona el borde de bajada
- cambia la autocorrelación y reduce el ancho de banda efectivo

---

# 6. Demodulación coherente

Con error de fase $\phi$:

$$
\widehat{m}_{\mathrm{coh}}[n]\approx m[n]\cos\phi + n_b[n]
$$

---

## Consecuencias en segundo orden

$$
R_{\widehat{m}_{\mathrm{coh}}}[\ell]\approx \cos^2\phi\,R_m[\ell]+R_{n_b}[\ell]
$$

$$
S_{\widehat{m}_{\mathrm{coh}}}(f)\approx \cos^2\phi\,S_m(f)+S_{n_b}(f)
$$

---

## Interpretación

- la estructura del ACF se preserva escalada
- el ruido se suma en segundo orden
- el error de fase reduce la potencia recuperada

---

# 7. Demodulación IQ

El receptor IQ estima la envolvente compleja:

$$
\widehat{u}[n]\approx A_c m[n]+n_I[n]+j n_Q[n]
$$

(en DSB-SC) o

$$
\widehat{u}[n]\approx A_c[1+\mu m[n]]+n_I[n]+j n_Q[n]
$$

(en DSB-LC)

---

## Interpretación

- el ruido afecta independientemente I y Q
- la representación compleja preserva la estructura del mensaje
- el error de fase se traduce en rotación del plano complejo
- el IQ imbalance introduce mezcla entre cuadraturas

---

## Efectos en PSD y ACF

- distorsión del espectro por fuga entre canales I/Q
- cambios en la simetría del ACF
- degradación incluso si la fase está corregida

---

# Conclusión general

El análisis de ACF, PSD y ancho de banda muestra que:

- el pulso rectangular domina el comportamiento espectral base
- la modulación desplaza o replica dicho espectro sin cambiar su estructura fundamental
- el ruido AWGN modifica aditivamente el segundo orden
- los demoduladores afectan principalmente:
  - escala (coherente)
  - valor absoluto (envolvente en SC)
  - geometría compleja (IQ)

En conjunto, toda la cadena preserva la estructura del mensaje $m[n]$, pero con degradaciones distintas según la etapa del sistema.


### Interpretación empírica

Las tablas y gráficas deben interpretarse en conjunto con los modelos analíticos de cada receptor.

---

- La señal de mensaje es un pulso rectangular unipolar de amplitud 1 y duración de 2048 muestras dentro de un registro de 4096 muestras. Sus bordes abruptos generan lóbulos laterales tipo sinc, por lo que el ancho de banda empírico depende del criterio de energía ocupada y de si se elimina o no el componente DC.

---

- Para DSB-LC con esta señal rectangular unipolar, todos los índices de modulación considerados mantienen la condición $1+\mu m[n]\geq 0$, por lo que la envolvente no se invierte. El aumento de $\mu$ se traduce principalmente en mayor profundidad de envolvente y mayor potencia en las bandas laterales.

---

- Para DSB-SC con esta señal no negativa, la detección por envolvente puede recuperar el soporte del pulso en el caso ideal sin ruido. El error práctico está dominado por el ruido AWGN y por la inercia del detector, más que por pérdida de signo (ya que el mensaje no es bipolar).

---

- Una inercia elevada en el detector de envolvente suaviza excesivamente la señal. Esto reduce el rizado de la portadora, pero también alarga el flanco de bajada del pulso rectangular, lo que altera la autocorrelación (ACF), la PSD y el error en forma de onda.

---

- La demodulación coherente degrada según el factor analítico $\cos\phi$. La forma de onda puede mantenerse similar tras una calibración afín, pero el NMSE aumenta porque el receptor diseñado para fase cero recibe una señal atenuada y, en DSB-LC, además un nivel DC escalado incorrectamente.

---

- La demodulación IQ permite corregir rotaciones puras de fase mediante una rotación del plano complejo de la envolvente. Sin embargo, el desbalance IQ es distinto: ganancias desiguales entre ramas I/Q y errores de cuadratura deforman la geometría del plano complejo, por lo que la corrección de fase por sí sola no elimina completamente el error residual.
