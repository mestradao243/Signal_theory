# Parte 1 — Verificación de la función de autocorrelación

Considere la función:

$$
R_x(\lambda)
=
\sum_{k=-K}^{K}
W_k \, e^{-\alpha_k \lambda^2}
$$

donde:

- $W_k \ge 0$
- $\alpha_k > 0$
- $\lambda$ es el desplazamiento temporal
- $K$ es un entero positivo

---

## 1. Verificación de simetría

Una función de autocorrelación válida debe cumplir:

$$
R_x(\lambda) = R_x(-\lambda)
$$

Evaluamos:

$$
R_x(-\lambda)
=
\sum_{k=-K}^{K}
W_k \, e^{-\alpha_k (-\lambda)^2}
$$

Como:

$$
(-\lambda)^2 = \lambda^2
$$

entonces:

$$
R_x(-\lambda)
=
\sum_{k=-K}^{K}
W_k \, e^{-\alpha_k \lambda^2}
$$

Por lo tanto,

$$
R_x(-\lambda) = R_x(\lambda)
$$

✔ La función es **simétrica**

---

## 2. Verificación de valor máximo en $\lambda = 0$

Evaluamos la función en $\lambda = 0$:

$$
R_x(0)
=
\sum_{k=-K}^{K}
W_k \, e^{-\alpha_k (0)^2}
$$

$$
R_x(0)
=
\sum_{k=-K}^{K}
W_k
$$

Para cualquier $\lambda \neq 0$:

$$
e^{-\alpha_k \lambda^2} < 1
$$

porque:

$$
\alpha_k > 0
$$

Por lo tanto:

$$
R_x(\lambda)
<
\sum_{k=-K}^{K}
W_k
=
R_x(0)
$$

✔ El valor máximo ocurre en $\lambda = 0$

---

## 3. Verificación de positividad definida

Una función es definida positiva si su transformada de Fourier es no negativa.

Primero observamos que cada término:

$$
W_k e^{-\alpha_k \lambda^2}
$$

es:

- positivo o cero, porque $W_k \ge 0$
- una función gaussiana, que es definida positiva

Una suma de funciones definidas positivas con coeficientes no negativos también es definida positiva.

Por lo tanto,

$$
R_x(\lambda)
$$

es una función definida positiva.

✔ La función cumple la propiedad de positividad definida.

---

## 4. Cálculo de la transformada de Fourier

La transformada de Fourier de una gaussiana es otra gaussiana:

$$
\mathcal{F}
\left\{
e^{-\alpha_k \lambda^2}
\right\}
=
\sqrt{\frac{\pi}{\alpha_k}}
\,
e^{-\frac{\pi^2 f^2}{\alpha_k}}
$$

Aplicando la propiedad de linealidad:

$$
S_x(f)
=
\mathcal{F}
\{R_x(\lambda)\}
$$

$$
=
\sum_{k=-K}^{K}
W_k
\sqrt{\frac{\pi}{\alpha_k}}
\,
e^{-\frac{\pi^2 f^2}{\alpha_k}}
$$

---

## 5. Verificación de no negatividad de la PSD

Observamos que cada término:

$$
W_k \ge 0
$$

$$
\sqrt{\frac{\pi}{\alpha_k}} > 0
$$

$$
e^{-\frac{\pi^2 f^2}{\alpha_k}} > 0
$$

Por lo tanto:

$$
S_x(f) \ge 0 \quad \forall f
$$

✔ La densidad espectral de potencia es no negativa.

---

## 6. Conclusión

La función:

$$
R_x(\lambda)
=
\sum_{k=-K}^{K}
W_k \, e^{-\alpha_k \lambda^2}
$$

cumple todas las propiedades requeridas para ser una función de autocorrelación válida:

- ✔ Es simétrica  
- ✔ Tiene su valor máximo en $\lambda = 0$  
- ✔ Es definida positiva  
- ✔ Su transformada de Fourier existe  
- ✔ Su densidad espectral de potencia es no negativa  

---

## Conclusión final

$$
\boxed{
R_x(\lambda)
\text{ es una función de autocorrelación válida de un proceso estacionario en sentido amplio (WSS).}
}
$$

# Parte 2 — Obtención y análisis de la densidad espectral de potencia

```python
# ============================================
# Parte 2 — Obtención y análisis de la PSD
# ============================================

# Librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift, fftfreq
from scipy.integrate import simpson

# Configuración de gráficos
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['font.size'] = 12

# ============================================
# Definición de parámetros
# ============================================

K = 3

# pesos (>=0)
W = np.array([0.4, 0.8, 1.0, 0.8, 0.4, 0.3, 0.2])

# alphas (>0)
alpha = np.array([0.5, 1.0, 2.0, 3.0, 2.0, 1.0, 0.5])

# índice k
k_vals = np.arange(-K, K+1)

# eje temporal
lambda_max = 5
N = 2000
lam = np.linspace(-lambda_max, lambda_max, N)

# ============================================
# Función de autocorrelación
# ============================================

def autocorrelation(lam, W, alpha):
    R = np.zeros_like(lam)

    for i in range(len(W)):
        R += W[i] * np.exp(-alpha[i] * lam**2)

    return R

R = autocorrelation(lam, W, alpha)

# ============================================
# PSD analítica
# ============================================

def psd_analytic(f, W, alpha):
    S = np.zeros_like(f)

    for i in range(len(W)):
        S += W[i] * np.sqrt(np.pi / alpha[i]) * np.exp(-(np.pi**2 * f**2) / alpha[i])

    return S

# ============================================
# PSD numérica usando FFT
# ============================================

dlam = lam[1] - lam[0]

S_num = fftshift(np.real(fft(R))) * dlam
freq = fftshift(fftfreq(N, dlam))

# PSD analítica en mismo eje
S_an = psd_analytic(freq, W, alpha)

# ============================================
# Gráfica autocorrelación
# ============================================

plt.figure()
plt.plot(lam, R, linewidth=2)
plt.title("Función de Autocorrelación")
plt.xlabel(r"$\lambda$")
plt.ylabel(r"$R_x(\lambda)$")
plt.grid()
plt.show()

# ============================================
# Gráfica PSD numérica
# ============================================

plt.figure()
plt.plot(freq, S_num, linewidth=2)
plt.title("PSD numérica (FFT)")
plt.xlabel("Frecuencia")
plt.ylabel(r"$S_x(f)$")
plt.grid()
plt.show()

# ============================================
# Gráfica PSD analítica
# ============================================

plt.figure()
plt.plot(freq, S_an, linewidth=2)
plt.title("PSD analítica")
plt.xlabel("Frecuencia")
plt.ylabel(r"$S_x(f)$")
plt.grid()
plt.show()

# ============================================
# Comparación PSD analítica vs numérica
# ============================================

plt.figure()
plt.plot(freq, S_num, label="PSD numérica", linewidth=2)
plt.plot(freq, S_an, '--', label="PSD analítica", linewidth=2)
plt.title("Comparación PSD analítica vs numérica")
plt.xlabel("Frecuencia")
plt.ylabel(r"$S_x(f)$")
plt.legend()
plt.grid()
plt.show()

# ============================================
# Análisis ancho temporal vs ancho espectral
# ============================================

# ancho temporal (varianza equivalente)
mean_lam = simpson(lam * R, lam) / simpson(R, lam)
var_lam = simpson((lam - mean_lam)**2 * R, lam) / simpson(R, lam)

# ancho espectral
mean_f = simpson(freq * S_an, freq) / simpson(S_an, freq)
var_f = simpson((freq - mean_f)**2 * S_an, freq) / simpson(S_an, freq)

print("Ancho temporal (varianza):", var_lam)
print("Ancho espectral (varianza):", var_f)

print("\nRelación inversa observada:")
print("Si la autocorrelación es más angosta → la PSD es más ancha")
print("Si la autocorrelación es más ancha → la PSD es más angosta")
```

# Parte 3 — Generación de procesos estocásticos

```python
# ============================================
# Parte 3 — Generación de procesos estocásticos
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter

plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['font.size'] = 12

# ============================================
# Parámetros generales
# ============================================

N = 5000          # número de muestras
mu = 0            # media
sigma = 1         # desviación estándar

np.random.seed(0)  # reproducibilidad

# ============================================
# 11. Ruido blanco gaussiano
# ============================================

white_noise = np.random.normal(mu, sigma, N)

# ============================================
# 12. Proceso ARMA(p,q)
# ============================================

# parámetros ARMA(2,1)
# AR: x[n] - 0.75x[n-1] + 0.25x[n-2]
# MA: w[n] + 0.65w[n-1]

a = [1, -0.75, 0.25]   # coeficientes AR
b = [1, 0.65]          # coeficientes MA

w = np.random.normal(mu, sigma, N)

arma_process = lfilter(b, a, w)

# ============================================
# 13. Random Walk
# ============================================

w_rw = np.random.normal(mu, sigma, N)

random_walk = np.zeros(N)

for i in range(1, N):
    random_walk[i] = random_walk[i-1] + w_rw[i]

# ============================================
# Gráficas de los procesos
# ============================================

# Ruido blanco
plt.figure()
plt.plot(white_noise)
plt.title("Ruido blanco gaussiano")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid()
plt.show()

# ARMA
plt.figure()
plt.plot(arma_process)
plt.title("Proceso ARMA(2,1)")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid()
plt.show()

# Random walk
plt.figure()
plt.plot(random_walk)
plt.title("Random Walk")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid()
plt.show()

# ============================================
# Mostrar estadísticas básicas
# ============================================

print("Ruido blanco:")
print("Media:", np.mean(white_noise))
print("Varianza:", np.var(white_noise))

print("\nARMA:")
print("Media:", np.mean(arma_process))
print("Varianza:", np.var(arma_process))

print("\nRandom Walk:")
print("Media:", np.mean(random_walk))
print("Varianza:", np.var(random_walk))
```

```python
# ============================================
# Parte 4 — Análisis en el dominio del tiempo
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# Usamos los procesos ya generados:
# white_noise
# arma_process
# random_walk

# ============================================
# Función para analizar media y varianza temporal
# ============================================

def analizar_proceso(x, nombre):

    N = len(x)

    # media global
    media = np.mean(x)

    # varianza global
    varianza = np.var(x)

    print(f"\n==== {nombre} ====")
    print(f"Media temporal: {media:.5f}")
    print(f"Varianza temporal: {varianza:.5f}")

    # ============================================
    # Media acumulada
    # ============================================

    media_acumulada = np.cumsum(x) / np.arange(1, N+1)

    # ============================================
    # Varianza acumulada
    # ============================================

    varianza_acumulada = np.zeros(N)

    for i in range(N):
        varianza_acumulada[i] = np.var(x[:i+1])

    # ============================================
    # Gráfica de la señal
    # ============================================

    plt.figure()
    plt.plot(x)
    plt.title(f"{nombre} — Realización temporal")
    plt.xlabel("n")
    plt.ylabel("x[n]")
    plt.grid()
    plt.show()

    # ============================================
    # Gráfica media acumulada
    # ============================================

    plt.figure()
    plt.plot(media_acumulada)
    plt.title(f"{nombre} — Media acumulada")
    plt.xlabel("n")
    plt.ylabel("Media")
    plt.grid()
    plt.show()

    # ============================================
    # Gráfica varianza acumulada
    # ============================================

    plt.figure()
    plt.plot(varianza_acumulada)
    plt.title(f"{nombre} — Varianza acumulada")
    plt.xlabel("n")
    plt.ylabel("Varianza")
    plt.grid()
    plt.show()


# ============================================
# Analizar cada proceso
# ============================================

analizar_proceso(white_noise, "Ruido blanco gaussiano")

analizar_proceso(arma_process, "Proceso ARMA(2,1)")

analizar_proceso(random_walk, "Random Walk")
```

# Parte 5 — Cálculo y análisis de la autocorrelación experimental

```python
# ============================================
# Parte 5 — Autocorrelación experimental
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# ============================================
# Función para calcular autocorrelación experimental
# ============================================

def autocorrelacion_experimental(x, max_lag):

    N = len(x)
    R = np.zeros(max_lag)

    for lag in range(max_lag):

        suma = 0

        for n in range(N - lag):
            suma += x[n] * x[n + lag]

        R[lag] = suma / N

    return R


# ============================================
# Parámetro de análisis
# ============================================

max_lag = 200


# ============================================
# Calcular autocorrelaciones
# ============================================

R_white = autocorrelacion_experimental(white_noise, max_lag)

R_arma = autocorrelacion_experimental(arma_process, max_lag)

R_rw = autocorrelacion_experimental(random_walk, max_lag)


# ============================================
# Graficar resultados
# ============================================

lags = np.arange(max_lag)

plt.figure()
plt.plot(lags, R_white)
plt.title("Autocorrelación experimental — Ruido blanco")
plt.xlabel("Lag")
plt.ylabel("R̂x(lag)")
plt.grid()
plt.show()


plt.figure()
plt.plot(lags, R_arma)
plt.title("Autocorrelación experimental — ARMA(2,1)")
plt.xlabel("Lag")
plt.ylabel("R̂x(lag)")
plt.grid()
plt.show()


plt.figure()
plt.plot(lags, R_rw)
plt.title("Autocorrelación experimental — Random Walk")
plt.xlabel("Lag")
plt.ylabel("R̂x(lag)")
plt.grid()
plt.show()
```

# Parte 6 — Cálculo y análisis de la densidad espectral de potencia

```python
# ============================================
# Parte 6 — Densidad espectral de potencia (PSD)
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# ============================================
# Función para calcular PSD usando FFT
# ============================================

def calcular_psd(x):

    N = len(x)

    # FFT
    X = np.fft.fft(x)

    # PSD
    PSD = (1/N) * np.abs(X)**2

    # frecuencias
    frecuencias = np.fft.fftfreq(N, d=1)

    # ordenar para graficar correctamente
    idx = np.argsort(frecuencias)

    return frecuencias[idx], PSD[idx]


# ============================================
# Calcular PSD de cada proceso
# ============================================

f_white, PSD_white = calcular_psd(white_noise)

f_arma, PSD_arma = calcular_psd(arma_process)

f_rw, PSD_rw = calcular_psd(random_walk)


# ============================================
# Graficar PSD — Ruido blanco
# ============================================

plt.figure()
plt.plot(f_white, PSD_white)
plt.title("PSD — Ruido blanco gaussiano")
plt.xlabel("Frecuencia")
plt.ylabel("PSD")
plt.grid()
plt.show()


# ============================================
# Graficar PSD — ARMA
# ============================================

plt.figure()
plt.plot(f_arma, PSD_arma)
plt.title("PSD — Proceso ARMA(2,1)")
plt.xlabel("Frecuencia")
plt.ylabel("PSD")
plt.grid()
plt.show()


# ============================================
# Graficar PSD — Random Walk
# ============================================

plt.figure()
plt.plot(f_rw, PSD_rw)
plt.title("PSD — Random Walk")
plt.xlabel("Frecuencia")
plt.ylabel("PSD")
plt.grid()
plt.show()
```

# Parte 7 — Verificación de estacionariedad

```python
# ============================================
# Parte 7 — Verificación de estacionariedad
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# ============================================
# Función para analizar estacionariedad
# ============================================

def verificar_estacionariedad(x, nombre):

    N = len(x)

    # dividir señal en segmentos
    segmentos = 10
    tamaño = N // segmentos

    medias = []
    varianzas = []

    for i in range(segmentos):

        inicio = i * tamaño
        fin = inicio + tamaño

        segmento = x[inicio:fin]

        medias.append(np.mean(segmento))
        varianzas.append(np.var(segmento))

    medias = np.array(medias)
    varianzas = np.array(varianzas)

    print(f"\n==== {nombre} ====")
    print("Media por segmentos:")
    print(medias)

    print("\nVarianza por segmentos:")
    print(varianzas)

    print("\nDesviación estándar de la media:", np.std(medias))
    print("Desviación estándar de la varianza:", np.std(varianzas))

    # graficar medias
    plt.figure()
    plt.plot(medias, marker='o')
    plt.title(f"{nombre} — Media por segmentos")
    plt.xlabel("Segmento")
    plt.ylabel("Media")
    plt.grid()
    plt.show()

    # graficar varianzas
    plt.figure()
    plt.plot(varianzas, marker='o')
    plt.title(f"{nombre} — Varianza por segmentos")
    plt.xlabel("Segmento")
    plt.ylabel("Varianza")
    plt.grid()
    plt.show()

    # criterio simple
    if np.std(medias) < 0.1 and np.std(varianzas) < 0.1:
        print("\nConclusión: Aproximadamente estacionario (WSS)")
    else:
        print("\nConclusión: No estacionario")


# ============================================
# Aplicar a cada proceso
# ============================================

verificar_estacionariedad(white_noise, "Ruido blanco")

verificar_estacionariedad(arma_process, "Proceso ARMA(2,1)")

verificar_estacionariedad(random_walk, "Random Walk")
```

# Parte 8 — Verificación de ergodicidad

```python
# ============================================
# Parte 8 — Verificación de ergodicidad
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter

# ============================================
# Parámetros
# ============================================

N = 5000
M = 50   # número de realizaciones

mu = 0
sigma = 1

# coeficientes ARMA
a = [1, -0.75, 0.25]
b = [1, 0.65]


# ============================================
# Funciones generadoras
# ============================================

def generar_ruido_blanco():
    return np.random.normal(mu, sigma, N)

def generar_arma():
    w = np.random.normal(mu, sigma, N)
    return lfilter(b, a, w)

def generar_random_walk():
    w = np.random.normal(mu, sigma, N)
    x = np.zeros(N)
    for i in range(1, N):
        x[i] = x[i-1] + w[i]
    return x


# ============================================
# Función para verificar ergodicidad
# ============================================

def verificar_ergodicidad(generador, nombre):

    medias_temporales = []
    varianzas_temporales = []

    realizaciones = []

    for i in range(M):

        x = generador()

        realizaciones.append(x)

        medias_temporales.append(np.mean(x))
        varianzas_temporales.append(np.var(x))

    medias_temporales = np.array(medias_temporales)
    varianzas_temporales = np.array(varianzas_temporales)

    # media estadística
    media_estadistica = np.mean(medias_temporales)

    # varianza estadística
    varianza_estadistica = np.mean(varianzas_temporales)

    print(f"\n==== {nombre} ====")

    print("Media temporal promedio:", media_estadistica)
    print("Varianza temporal promedio:", varianza_estadistica)

    print("Desviación de medias:", np.std(medias_temporales))
    print("Desviación de varianzas:", np.std(varianzas_temporales))

    # graficar medias
    plt.figure()
    plt.plot(medias_temporales, marker='o')
    plt.title(f"{nombre} — Medias de múltiples realizaciones")
    plt.xlabel("Realización")
    plt.ylabel("Media temporal")
    plt.grid()
    plt.show()

    # graficar varianzas
    plt.figure()
    plt.plot(varianzas_temporales, marker='o')
    plt.title(f"{nombre} — Varianzas de múltiples realizaciones")
    plt.xlabel("Realización")
    plt.ylabel("Varianza temporal")
    plt.grid()
    plt.show()

    # criterio
    if np.std(medias_temporales) < 0.1:
        print("Conclusión: Ergodico en media")
    else:
        print("Conclusión: No ergodico en media")


# ============================================
# Ejecutar análisis
# ============================================

verificar_ergodicidad(generar_ruido_blanco, "Ruido blanco")

verificar_ergodicidad(generar_arma, "Proceso ARMA")

verificar_ergodicidad(generar_random_walk, "Random Walk")
```

# Parte 9 — Determinación del número de componentes gaussianas $K$

```python
# ============================================
# Parte 9 — Análisis del parámetro K
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# ============================================
# Dominio temporal
# ============================================

lambda_vals = np.linspace(-5, 5, 2000)

# ============================================
# Función de autocorrelación teórica
# ============================================

def autocorrelacion_teorica(lambda_vals, K):

    R = np.zeros_like(lambda_vals)

    for k in range(-K, K+1):

        # pesos positivos
        Wk = np.exp(-0.3 * abs(k))

        # parámetros positivos
        alpha_k = 0.5 + 0.2 * abs(k)

        R += Wk * np.exp(-alpha_k * lambda_vals**2)

    return R


# ============================================
# Función para calcular PSD usando FFT
# ============================================

def calcular_psd(R, lambda_vals):

    N = len(R)

    delta = lambda_vals[1] - lambda_vals[0]

    PSD = np.abs(np.fft.fft(R)) * delta

    freq = np.fft.fftfreq(N, delta)

    idx = np.argsort(freq)

    return freq[idx], PSD[idx]


# ============================================
# Valores de K a analizar
# ============================================

K_values = [0, 1, 2, 5, 10, 20]

# ============================================
# Graficar autocorrelación para distintos K
# ============================================

plt.figure()

for K in K_values:

    R = autocorrelacion_teorica(lambda_vals, K)

    plt.plot(lambda_vals, R, label=f"K = {K}")

plt.title("Autocorrelación para distintos valores de K")
plt.xlabel("λ")
plt.ylabel("R_x(λ)")
plt.legend()
plt.grid()
plt.show()


# ============================================
# Graficar PSD para distintos K
# ============================================

plt.figure()

for K in K_values:

    R = autocorrelacion_teorica(lambda_vals, K)

    freq, PSD = calcular_psd(R, lambda_vals)

    plt.plot(freq, PSD, label=f"K = {K}")

plt.title("PSD para distintos valores de K")
plt.xlabel("Frecuencia")
plt.ylabel("S_x(f)")
plt.legend()
plt.grid()
plt.show()


# ============================================
# Verificar estabilidad numérica
# ============================================

print("Análisis de estabilidad:\n")

for K in K_values:

    R = autocorrelacion_teorica(lambda_vals, K)

    print(f"K = {K}")
    print("Valor máximo:", np.max(R))
    print("Valor mínimo:", np.min(R))
    print("¿Es positiva definida aproximada?:", np.min(R) >= 0)
    print()
```

# Parte 10 — Análisis comparativo final

## 10.1 Comparación de estacionariedad

Se analizó la estacionariedad en sentido amplio (WSS) de los procesos generados: ruido blanco gaussiano, proceso ARMA(p,q) y caminata aleatoria (Random Walk). Para que un proceso sea estacionario en sentido amplio, debe cumplir las siguientes condiciones:

- Media constante en el tiempo
- Varianza constante en el tiempo
- Autocorrelación dependiente únicamente del desplazamiento temporal

### Ruido blanco gaussiano

El ruido blanco gaussiano presentó una media aproximadamente constante e igual a cero en todos los segmentos analizados. Asimismo, la varianza se mantuvo constante en el tiempo. La función de autocorrelación mostró un comportamiento tipo delta, indicando que depende únicamente del desplazamiento y no del tiempo absoluto.

Por lo tanto, el ruido blanco gaussiano es estacionario en sentido amplio.

### Proceso ARMA(p,q)

El proceso ARMA presentó una media constante en el tiempo, con pequeñas fluctuaciones debidas a la naturaleza aleatoria del proceso. La varianza también se mantuvo aproximadamente constante. La autocorrelación mostró un decaimiento progresivo en función del desplazamiento, lo cual confirma que depende únicamente de la diferencia temporal.

Por lo tanto, el proceso ARMA(p,q) es estacionario en sentido amplio.

### Caminata aleatoria (Random Walk)

La caminata aleatoria mostró una media que varía significativamente con el tiempo. Además, la varianza aumenta progresivamente, lo cual es característico de este tipo de proceso. La autocorrelación no depende únicamente del desplazamiento, sino también del tiempo absoluto.

Por lo tanto, la caminata aleatoria no es estacionaria en sentido amplio.

---

## 10.2 Comparación de ergodicidad

La ergodicidad implica que las propiedades estadísticas de un proceso pueden obtenerse a partir de una sola realización suficientemente larga, es decir, que el promedio temporal es igual al promedio estadístico entre múltiples realizaciones.

### Ruido blanco gaussiano

Se observó que la media temporal de múltiples realizaciones converge al mismo valor que la media estadística. Asimismo, la varianza también converge a un valor constante.

Por lo tanto, el ruido blanco gaussiano es ergódico en media y en autocorrelación.

### Proceso ARMA(p,q)

El proceso ARMA mostró un comportamiento similar al ruido blanco, con medias temporales y estadísticas consistentes entre sí. La varianza también mostró convergencia entre realizaciones.

Por lo tanto, el proceso ARMA(p,q) es ergódico en media y en autocorrelación.

### Caminata aleatoria

La caminata aleatoria mostró variaciones significativas en la media entre distintas realizaciones, y la varianza crece con el tiempo. Esto indica que las propiedades estadísticas no pueden obtenerse a partir de una única realización.

Por lo tanto, la caminata aleatoria no es ergódica.

---

## 10.3 Comparación de las funciones de autocorrelación

La función de autocorrelación describe la dependencia temporal del proceso.

- El ruido blanco gaussiano presenta una autocorrelación tipo delta, lo cual indica ausencia de memoria temporal.
- El proceso ARMA presenta una autocorrelación que decrece gradualmente, indicando memoria temporal finita.
- La caminata aleatoria presenta una autocorrelación que no decae adecuadamente, indicando memoria infinita y no estacionariedad.

Esto confirma el comportamiento teórico esperado para cada tipo de proceso.

---

## 10.4 Comparación de las densidades espectrales de potencia (PSD)

La densidad espectral de potencia describe la distribución de potencia del proceso en el dominio de la frecuencia.

- El ruido blanco gaussiano presenta una PSD aproximadamente constante, indicando que todas las frecuencias están presentes con igual potencia.
- El proceso ARMA presenta una PSD estructurada, lo cual indica que ciertas frecuencias dominan el comportamiento del proceso.
- La caminata aleatoria presenta una PSD concentrada en bajas frecuencias, indicando variaciones lentas en el tiempo.

Estos resultados son consistentes con el teorema de Wiener-Khinchin, el cual establece la relación entre la autocorrelación y la densidad espectral de potencia.

---

## 10.5 Conclusión general

A partir del análisis teórico y experimental realizado, se concluye lo siguiente:

- El ruido blanco gaussiano es estacionario en sentido amplio y ergódico.
- El proceso ARMA(p,q) es estacionario en sentido amplio y ergódico.
- La caminata aleatoria no es estacionaria ni ergódica.

Además, se verificó experimentalmente la relación entre la función de autocorrelación y la densidad espectral de potencia, confirmando el teorema de Wiener-Khinchin.

Los resultados obtenidos son consistentes con la teoría de procesos estocásticos y validan el uso de la función de autocorrelación como herramienta fundamental para el análisis de procesos aleatorios.
